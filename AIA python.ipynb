We understood the data,Handled missing values, fixed datatypes, removed duplicates and done a visual representation of the given data.

We are importing pandas and reading 'train.csv' using this command.

```python
import pandas as pd
data = pd.read_csv('train.csv')
```

Then we checked First and last details of the given data.

```python

print(data.head())
print(data.tail())
```

After that we checked information of the given Data.

```python
print(data.info())
```

Then we checked numerical discription of the given Data.

```python
print(data.describe())
```

Then we Removed the duplicates from the given data.

```python
data = (data.drop_duplicates())
```

Then we checked Null values from the data.

```python
print(data.isnull().sum())
```

We filled the Null values of Postal Code.

```python
data["Postal Code"] = data["Postal Code"].fillna(data["Postal Code"].median())
```

We checked Null values again!

```python
print(data.isnull().sum())
```

Then we imported matplotlib and seaborn.
After that we made a barplot for a visual representation of Sales Distribution by Categories.

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.barplot(y="Sales",x="Category", data=data)
plt.title("Sales Distribution by Categories")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()
```

Then we made a countplot for a visual representation of Segment details.

```python
sns.countplot(x="Ship Mode",hue="Segment", data=data)
plt.title("Segment details")
plt.xlabel("Ship Mode")
plt.ylabel("Segment")
plt.show()
```

Then we made a lineplot for a visual representation of Sales details in a Region.

```python
sns.lineplot(x="Region",y="Sales", data=data)
plt.title("Sales details in Region")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.show()
```

Then we made a pointplot for a visual representation of Phone Sales in a Region.

```python
sns.pointplot(x="Region",y="Sub-Category", data=data)
plt.title("Phone sales in Region")
plt.xlabel("Region")
plt.ylabel("Sub-Category")
plt.show()
```
